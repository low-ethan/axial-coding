```python
import pandas as pd
import ast
```

```python
high = pd.read_csv('good/gemma3_27b.csv')
low = pd.read_csv('good/gemma3_27_temp0.csv')
```

```python
df_full = pd.merge(high, low, on='id', suffixes=('_high', '_low'))
```

```python
df_full['json_high'] = df_full['json_high'].apply(ast.literal_eval)
df_full['json_low'] = df_full['json_low'].apply(ast.literal_eval)
```

```python
df_full['high_codes'] = df_full['json_high'].apply(lambda x: [tmp['Code'] for tmp in x])
df_full['low_codes'] = df_full['json_low'].apply(lambda x: [tmp['Code'] for tmp in x])
```

```python
df_full[df_full['high_codes'].astype(str) != df_full['low_codes'].astype(str)]
```

```python
def count_matches(row):
    high = row['high_codes']
    low = row['low_codes']
    matches = sum(h == l for h, l in zip(high, low))
    mismatches = len(high) - matches  # assuming both lists are same length
    return pd.Series({'num_matches': matches, 'num_mismatches': mismatches})

df_full[['num_matches', 'num_mismatches']] = df_full.apply(count_matches, axis=1)
```
